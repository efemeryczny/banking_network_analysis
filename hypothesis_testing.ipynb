```python
# Zmienna niezależna (In-degree)
X = data[['In-degree']]

# Zmienne zależne (miary ryzyka)
Y = data[['Beta', 'Debt-to-equity', 'ROA_stdev', 'ROE_stdev']]

# Dodaj stałą (intercept) do modelu
X = sm.add_constant(X)

# Przeprowadź analizę regresji osobno dla każdej zmiennej zależnej
for var in Y:
    Y = data[[var]]
    model = sm.OLS(Y, X).fit()
    print(f"Dependent Variable: {var}")
    print(model.summary())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Wczytanie danych z pliku Excel
excel_file_path = '/content/drive/MyDrive/Colab Notebooks/Google_drive/financials.xlsx'
data = pd.read_excel(excel_file_path)
data
```

```python
# Wybór odpowiednich kolumn jako zmienne niezależne (X) i zmienne zależne (Y)
X = data[['In-degree']]
Y = data[['Debt-to-equity', 'ROE_stdev', 'ROA_stdev']]

# Budowa modelu regresji wielorakiej
model = LinearRegression()
model.fit(X, Y)

# Predykcje
Y_pred = model.predict(X)

# Metryki oceny modelu (dla każdej zmiennej zależnej osobno)
r2_scores = r2_score(Y, Y_pred, multioutput='raw_values')
rmse_scores = mean_squared_error(Y, Y_pred, multioutput='raw_values', squared=False)

# Interpretacja wyników
coefficients = model.coef_
intercept = model.intercept_

print(f'Współczynniki regresji: {coefficients}')
print(f'Intercept: {intercept}')
print(f'R-squared (debt-to-equity): {r2_scores[0]}')
print(f'RMSE (debt-to-equity): {rmse_scores[0]}')
print(f'R-squared (ROE_std_dev): {r2_scores[1]}')
print(f'RMSE (ROE_std_dev): {rmse_scores[1]}')
print(f'R-squared (ROA_std_dev): {r2_scores[2]}')
print(f'RMSE (ROA_std_dev): {rmse_scores[2]}')
```

```python
# Define your dependent and independent variables
# Dependent variables (W)
Y = data[['Debt-to-equity', 'ROE_stdev', 'ROA_stdev']]

# Independent variables (ZS and ZKT)
X = data[['In-degree']]

# Add a constant term to the independent variables (intercept)
X = sm.add_constant(X)

# Perform multiple linear regression for each dependent variable
results = []
for col in Y.columns:
    model = sm.OLS(Y[col], X).fit()
    results.append(model)

# Print regression results
for i, col in enumerate(Y.columns):
    print(f"Regression results for {col}:")
    print(results[i].summary())

# Perform hypothesis tests (e.g., test the significance of coefficients)
# For example, to test if 'Degree' is a significant predictor for 'ROE':
hypothesis_test = results[0].t_test('Degree = 0')
print("\nHypothesis Test for Degree on ROE:")
print(hypothesis_test)
```
